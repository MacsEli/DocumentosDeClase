```python
url = 'https://www.rottentomatoes.com/top/bestofrt/'

html = requests.get(url).content
```

```python
soup = BeautifulSoup(html, 'lxml')

print(soup)
```

```python
tags = ['h1', 'h2', 'h3', 'p', 'a', 'href']

elements = [element.text for element in soup.find_all(tags)]
```

```python
elements
```

```python
# links = soup.select('td a[class="unstyled articleLink"]')
links = soup.select('td a.unstyled.articleLink')
```

```python
links
```

```python
title_clean = [element.text.strip() for element in links]
```

```python
title_clean
```

```python
reviews = soup.select('td[class="right hidden-xs"]')
```

```python
reviews_clean = [re.text for re in reviews]
reviews_clean
```

```python
links[0]['href']
```

```python
url = 'https://www.rottentomatoes.com'

links_to_movie = [url + links[i]['href'] for i in range(len(links))]

links_to_movie
```

```python
df = pd.DataFrame(list(zip(title_clean, reviews_clean, links_to_movie)),
                 columns = ['Movie', 'Reviews', 'Link'])
```
